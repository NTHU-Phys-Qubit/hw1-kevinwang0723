### Problem 1 1D Random walk

```python
import numpy as np
import matplotlib.pyplot as plt

N=5000
data=np.random.random((N,))-0.5
location=np.cumsum(data)

plt.hist(data,bins=20)
plt.title("distribution")
plt.show()
```

```python
# single move trajectory.
plt.title("Trajectory of Random Walk (-0.5~0.5) Distribution")
plt.ylabel('Location')
plt.plot(location)
plt.show()
```

```python
# unit move uniform distrbutre between (-0.5~0.5) for many trail and its final location. 
Trial=10000
final=np.zeros(Trial)
for j in range(0,Trial):
    data=np.random.random((N,))-0.5
    final[j]=np.sum(data)
```

```python
plt.hist(final,bins=50)
plt.title("Final Location of Random Walk (-0.5~0.5) Distribution")
plt.show()
```

```python
# unit move uniform distrbutre between (-1 or 1) for many trail and its final location. 
n_data=np.zeros(N)
for i in range(0,N):
    i
    if data[i]>0:
        n_data[i]=1
    else:
        n_data[i]=-1
plt.hist(n_data,bins=2)
```

```python
n_location=np.cumsum(n_data)
import matplotlib.pyplot as plt
plt.ylabel('n_Location')
plt.plot(n_location)
plt.title("Trajectory of Random Walk (-0.5 or 0.5) Distribution")
plt.show()
```

```python
Trial=5000
N=5000
final=np.zeros(Trial)
for j in range(0,Trial):
    n_data=np.zeros(N)
    data=np.random.random((N,))-0.5
    for i in range(0,N):
        if data[i]>0:
            n_data[i]=1
        else:
            n_data[i]=-1
    final[j]=np.sum(n_data)    
plt.hist(final,bins=30)
plt.title("Final Location of Random Walk (-0.5 or 0.5) Distribution")
plt.show()
```

### Problem 2

## random distribute from -1/2~1/2

```python
N=5000
xdata=2*(np.random.random((N,))-0.5)
ydata=2*(np.random.random((N,))-0.5)
xdata
xlocation=np.cumsum(xdata)
ylocation=np.cumsum(ydata)
plt.ylabel('n_Location')
plt.plot(xlocation,ylocation)
axes = plt.gca()
axes.set_xlim([-30,30])
axes.set_ylim([-30,30])
```

```python
# unit move uniform distrbutre between (-0.5~0.5) for many trail and its final location. 
Trial=10000
final_x=np.zeros(Trial)
final_y=np.zeros(Trial)
for j in range(0,Trial):
    data=np.random.random((N,))-0.5
    final_x[j]=np.sum(data)
    data=np.random.random((N,))-0.5
    final_y[j]=np.sum(data)
```

```python
plt.plot(final_x,final_y,'bo',markersize=3)
axes = plt.gca()
axes.set_xlim([-100,100])
axes.set_ylim([-100,100])
plt.show()
```

# Distance from orign (-1/2~1/2)

```python
r=np.zeros(N)
for i in range(0,N):
    r[i]=np.sqrt(final_x[i]*final_x[i]+final_y[i]*final_y[i])
plt.hist(r,bins=100)
plt.title("Histogram of 2D random walk final distance (-1/2~1/2) ")
plt.show()
```

## Random walk from orign (-1 or 1)

```python
N=5000
n_xdata=np.zeros(N)
n_ydata=np.zeros(N)

for i in range(0,N):
    data=np.random.random((N,))-0.5
    if data[i]>0:
        n_xdata[i]=1
    else:
        n_xdata[i]=-1
    data=np.random.random((N,))-0.5
    if data[i]>0:
        n_ydata[i]=1
    else:
        n_ydata[i]=-1
plt.hist(n_data,bins=2)
```

```python
x_location=np.cumsum(n_xdata)
y_location=np.cumsum(n_ydata)
plt.ylabel('n_Location')
plt.plot(x_location,y_location)
axes = plt.gca()
axes.set_xlim([-30,30])
axes.set_ylim([-30,30])
```

```python
# unit move uniform distrbutre between (-0.5 or 0.5) for many trail and its final location. 
Trial=10000
nfinal_x=np.zeros(Trial)
nfinal_y=np.zeros(Trial)
for j in range(0,Trial):
    
    data=np.random.random((N,))-0.5
    if data[i]>0:
        n_xdata[i]=1
    else:
        n_xdata[i]=-1
    data=np.random.random((N,))-0.5
    if data[i]>0:
        n_ydata[i]=1
    else:
        n_ydata[i]=-1
    nfinal_x[j]=np.sum(n_xdata)
    nfinal_y[j]=np.sum(n_ydata)
```

```python
plt.plot(nfinal_x,nfinal_y,'bo',markersize=3)
axes = plt.gca()
axes.set_xlim([-100,100])
axes.set_ylim([-100,100])
plt.show()
```

```python
r=np.zeros(N)
for i in range(0,N):
    r[i]=np.sqrt(nfinal_x[i]*nfinal_x[i]+nfinal_y[i]*nfinal_y[i])
plt.hist(r,bins=100)
plt.title("Histogram of 2D random walk final distance (-1/2 or 1/2) ")
plt.show()
```

## cos sin from [0 2pi]

```python
xdata=np.sin(2*np.pi*(np.random.random((N,))))
ydata=np.cos(2*np.pi*(np.random.random((N,))))
xlocation=np.cumsum(xdata)
ylocation=np.cumsum(ydata)
plt.ylabel('n_Location')
plt.plot(xlocation,ylocation)
axes = plt.gca()
axes.set_xlim([-30,30])
axes.set_ylim([-30,30])
```

```python
Trial=10000
final_x=np.zeros(Trial)
final_y=np.zeros(Trial)
for j in range(0,Trial):
    xdata=np.sin(2*np.pi*(np.random.random((N,))))
    final_x[j]=np.sum(xdata)
    ydata=np.cos(2*np.pi*(np.random.random((N,))))
    final_y[j]=np.sum(ydata)
```

```python
plt.plot(final_x,final_y,'bo',markersize=3)
plt.show()
```

```python
r=np.zeros(N)
for i in range(0,N):
    r[i]=np.sqrt(final_x[i]*final_x[i]+final_y[i]*final_y[i])
plt.hist(r,bins=100)
plt.title("Histogram of 2D random walk final distance (-1/2~1/2) ")
plt.show()
```
